Importing files


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
#Transfering our csv file into pd dataframe
card_data = pd.read_csv('/content/creditcard.csv')
```

```python
card_data.head()
```

```python
card_data.shape
```

```python
card_data.tail()
```

```python
card_data.info()
```

```python
card_data.isnull().sum()
```

```python
card_data['Class'].value_counts()
```

```python
#highly ununiform dataset as it contains more number of zeroes
```

```python
legal = card_data[card_data.Class == 0]
```

```python
fraud = card_data[card_data.Class == 1]
```

```python
legal.head()
```

```python
fraud.tail()
```

```python
legal.shape
```

```python
fraud.shape
```

```python
#making data uniform(same number of fraud and legal class)
```

```python
card_data.groupby('Class').mean()
```

```python
# legal_u is dataframe of legal dataset for uniform sampling of legal and fraud datasets
legal_u = legal.sample(492)
```

```python
legal_u.head()
```

```python
#now merging legl_u and fraud
dataset = pd.concat([legal_u, fraud], axis=0)
dataset.head()
```

```python
dataset.tail()
```

```python
dataset['Class'].value_counts()
```

```python
dataset.head()
```

```python
dataset.groupby('Class').mean()
```

```python
Y = dataset['Class']
```

```python
X = dataset.drop(columns = 'Class', axis = 1)
```

```python
#implementing model by splitting our refined sample and fitting it in suitable model
```

```python
Xtrain, Xtest, Ytrain, Ytest = train_test_split(X, Y, test_size=0.2, stratify = Y, random_state = 5)
```

```python
print(X.shape, Xtrain.shape,  Xtest.shape)
```

```python
Xtrain.head()
```

```python
#using logistic regression model
```

```python
md = LogisticRegression()
```

```python
#training LogisticRegression moddel by training data
```

```python
md.fit(Xtrain, Ytrain)
```

```python
#accuracy score of training data
```

```python
Ytrainpred = md.predict(Xtrain)
trainingaccuracy  = accuracy_score(Ytrainpred, Ytrain)
print('accuracy score of training data', trainingaccuracy )
```

```python
#accuracy score of test data
```

```python
Ytestpred = md.predict(Xtest)
trainingaccuracy  = accuracy_score(Ytestpred, Ytest)
print('accuracy score of test data', trainingaccuracy )
```

```python
#training KNN model by training data
```

```python
from sklearn.neighbors import KNeighborsClassifier
```

```python
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(Xtrain,Ytrain)
```

```python
#accuracy score of training data
```

```python
Ytrainpred = knn.predict(Xtrain)
trainingaccuracy  = accuracy_score(Ytrainpred, Ytrain)
print('accuracy score of training data', trainingaccuracy )
```

```python
#accuracy score of test data
```

```python
Ytestpred = knn.predict(Xtest)
trainingaccuracy  = accuracy_score(Ytestpred, Ytest)
print('accuracy score of test data', trainingaccuracy )
```

We got highest accuracy in LogisticRegression  model
so we will proceed with LogisticRegression  as our final model with accuracy of around 90%
